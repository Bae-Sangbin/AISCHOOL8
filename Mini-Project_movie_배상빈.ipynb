```python
df['개봉일'] = pd.to_datetime(df['개봉일'])

df['전국스크린수'] = df['전국\n스크린수'].str.replace(",", "")
df = df.drop(columns='전국\n스크린수')

df['전국매출액'] = df['전국\n매출액'].str.replace(",", "")
df = df.drop(columns='전국\n매출액')

df['서울매출액'] = df['서울\n매출액'].str.replace(",", "")
df = df.drop(columns='서울\n매출액')

df['전국관객수'] = df['전국\n관객수'].str.replace(",", "")
df = df.drop(columns='전국\n관객수')

df['서울관객수'] = df['서울\n관객수'].str.replace(",", "")
df = df.drop(columns='서울\n관객수')
```

```python
df['전국매출액']
```

```python
df['전국스크린수'] = df['전국스크린수'].astype('int64')
df['전국매출액'] = df['전국매출액'].astype('int64')
df['서울매출액'] = df['서울매출액'].astype('int64')
df['전국관객수'] = df['전국관객수'].astype('int64')
df['서울관객수'] = df['서울관객수'].astype('int64')
```

```python
df.dtypes
```

```python
df['전국매출액(천만)'] = round(df['전국매출액'] / 10000000,2)
df['서울매출액(천만)'] = round(df['서울매출액'] / 10000000,2)
df['전국관객수(만)'] = round(df['전국관객수'] / 10000,2)
df['서울관객수(만)'] = round(df['서울관객수'] / 10000,2)
```

```python
df = df.drop(columns='전국매출액')
df = df.drop(columns='서울매출액')
df = df.drop(columns='전국관객수')
df = df.drop(columns='서울관객수')
```
